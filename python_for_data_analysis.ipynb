```python
import pandas as pd

# Sample data
data = {
    'IP': ['A', 'A', 'A', 'A', 'B', 'B'],
    'URL ID': ['U1', 'U2', 'U12', 'U24', 'U1', 'U2'],
    'Sessions': [1, 1, 2, 2, 1, 1]
}

df = pd.DataFrame(data)

# Convert the "Sessions" column to a string
df["Sessions"] = df["Sessions"].astype(str)

# Group by 'IP' and aggregate 'URL ID' as a comma-separated string and sum 'Sessions'
transformed_df = df.groupby('IP').agg({'URL ID': ', '.join, 'Sessions': 'sum'}).reset_index()



# Display the transformed DataFrame
transformed_df
```
